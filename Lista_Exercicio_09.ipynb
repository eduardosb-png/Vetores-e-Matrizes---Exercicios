```python
#1 Faça um algoritmo que leia, mostre e calcule a soma dos elementos de uma matriz 7 X 9 de inteiros
matriz = []
soma = 0

for i in range(2):
  linha = []
  for j in range(2):
    valor = int(input(f"Digite o elemento da linha{i} e coluna{j}: "))
    linha.append(valor)
    soma = soma + valor
  matriz.append(linha)
print("\nMatriz digitada:")
for i in range(2):
  print(matriz[i])
print(f"\nA soma de todos os elementos é: {soma}")
```

```python
#2 Gerar (sem ler) e imprimir uma matriz de 10 linhas e 10 colunas cujos elementos são 1, 2, 3, ... , 99, 100.
matriz = []
for i in range(10):
  linha = []
  for j in range(1, 11):
    linha.append(i * 10 + j)
  matriz.append(linha)
for linha in matriz:
  print(linha)
```

```python
#3 Criar e imprimir a matriz identidade MI[1..7,1..7] em que todos os elementos da diagonal principal são iguais a 1 e os demais são zero.
matriz = []
for i in range(7):
  linha = []
  for j in range(7):
    if i == j:
      linha.append(1)
    else:
      linha.append(0)
  matriz.append(linha)

for linha in matriz:
  print(linha)
```

```python
#4 Dada uma matriz M[1..6,1..8], criar um vetor C que contenha, em cada posição, a quantidade de elementos negativos da linha correspondente de M.
matriz = []
for i in range(6):
  linha = []
  for j in range(8):
    valor = int(input(f"Digite o elemento da linha{i} e coluna{j}: "))
    linha.append(valor)
  matriz.append(linha)


print("Matriz M:")
for linha in matriz:
  print(linha)

C = []
for i in range(6):
  contador = 0
  for j in range(8):
    if matriz[i][j] < 0:
      contador += 1
  C.append(contador)

print("\nVetor C (negativos por linha):")
print(C)
```

```python
#5 Faça um algoritmo que leia uma matriz de 15 X 20 de reais e mostre a soma de cada coluna separadamente
matriz = []
for i in range(15):
  linha = []
  for j in range(20):
    valor = float(input(f"Digite o elemento da linha{i} e coluna{j}: "))
    linha.append(valor)
  matriz.append(linha)

S = []
for j in range(20):
  soma = 0
  for i in range(15):
    soma += matriz[i][j]
  S.append(soma)

print("\nSoma de cada coluna:")
for j in range(20):
  print(f"Coluna {j}: {S[j]}")
```

```python
#6 Dadas duas matrizes numéricas A, 5x5, e B, 5x5 criar e imprimir a matriz S, soma de A e B.
matrizA = []
for i in range(5):
  linha = []
  for j in range(5):
    valor = float(input(f"Digite o elemento da linha{i} e coluna{j} matrizA: "))
    linha.append(valor)
  matrizA.append(linha)

matrizB = []
for i in range(5):
  linha = []
  for j in range(5):
    valor = float(input(f"Digite o elemento da linha{i} e coluna{j} matrizB: "))
    linha.append(valor)
  matrizB.append(linha)

S = []
for i in range(5):
  linha = []
  for j in range(5):
    linha.append(matrizA[i][j] + matrizB[i][j])
  S.append(linha)

print("\nMatriz S (soma de A e B):")
for linha in S:
  print(linha)
```

```python
#7 Faça um algoritmo que calcule a média dos elementos da diagonal principal de uma matriz 10 X 10 de inteiros
matriz = []
for i in range(10):
  linha = []
  for j in range(10):
    valor = float(input(f"Digite o elemento da linha{i} e coluna{j}: "))
    linha.append(valor)
  matriz.append(linha)

soma = 0
for i in range(10):
  soma += matriz[i][i]

media = soma / 10

print("\nMedia da diagonal da matriz:")
print(media)
```

```python
#8 Faça um algoritmo que leia uma matriz 3x3 e após a leitura multiplique os elementos da diagonal principal com a média dos elementos da diagonal secundária.
matriz = []
for i in range(3):
  linha = []
  for j in range(3):
    valor = float(input(f"Digite o elemento da linha{i} e coluna{j}: "))
    linha.append(valor)
  matriz.append(linha)

soma = 0
for i in range(3):
  soma += matriz[i][2-i]
media = soma / 3

for i in range(3):
  matriz[i][i] = matriz[i][i] * media

print("\nMultiplicação da diagonal com a média da matriz:")
print(matriz)
```

```python
#9 Faça um algoritmo que leia uma matriz 50x50 de números reais. A seguir, multiplique cada linha pelo elemento da diagonal principal daquela linha. Mostre a matriz após as multiplicações.
matriz = []
for i in range(50):
    linha = []
    for j in range(50):
        valor = float(input(f"Digite o elemento da linha {i} e coluna {j}: "))
        linha.append(valor)
    matriz.append(linha)

for i in range(50):
    diagonal = matriz[i][i]
    for j in range(50):
        matriz[i][j] = matriz[i][j] * diagonal

print("\nMatriz após multiplicações:")
for linha in matriz:
  print(linha)
```
